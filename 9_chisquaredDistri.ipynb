```python
import numpy as np
```

```python
d1=np.random.chisquare(df=1,size=1000)
d2=np.random.chisquare(df=2,size=1000)
d3=np.random.chisquare(df=3,size=1000)
print(d1[:10])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

sns.kdeplot(d1, label='dof 1')
sns.kdeplot(d2, label='dof 2')
sns.kdeplot(d3,label='dof 3')
```

As degree of freedom increses -> chi square tends to normal distribution <br>

## Chi square test: <br>

To check if there is any relationship between two samples of data

```python
from scipy.stats import chi2_contingency
```

```python
data=[[10,20,30],[6,9,17]]
```

H0: there is no associatiation between variables
H1: there is a associatiation / correlation between the variables we are testing for

```python
stat,p_value,dof,chi_array =chi2_contingency(data)
p_value
```

p_value>0.05 so we reject null hypothesis H0
we accept the null hypothesis H1 i.e. there is a relationship between two variables 

##### Example2:

```python
data=[[10,20,30],[9,1,8]]
stat,p_value,dof,chi_array =chi2_contingency(data)
p_value
```

p_value<0.05 so we accept null hypothesis H0
i.e. there is no relationship between two variables

```python
import random
a=[random.gauss(50,20) for x in range(30)]
b=[random.gauss(55,15) for x in range(30)]
```

```python
import seaborn as sns

sns.set_style('darkgrid')
sns.kdeplot(a,fill=True)
sns.kdeplot(b,fill=True)
```

## Independent Samaple T-test

H0 : null Hypothesis : Mean A = Mean B

```python
#t test on data
import scipy.stats as stats
```

```python
t_stat,p_value=stats.ttest_ind(a,b,equal_var=False)
```

```python
t_stat,p_value
```

since the p-value>0.05 we accept the null hypothesis i.e. means of both the distributions are almost same

```python
import numpy as np
```

```python
np.mean(a), np.mean(b)
```

## Paired sample T-Test

Null Hypotheses H0 : Difference of two Means=0
Alternative Hypothesis H1:  Difference of two Means != 0

```python
# other sample
a=[random.gauss(50,20) for x in range(30)]
b=[random.gauss(60,25) for x in range(30)]
```

```python
t_stat,p_value=stats.ttest_rel(a,b)
```

```python
t_stat,p_value
```

since the p_value<0.05 we reject null hypothesis

Lets confirm that

```python
np.mean(b)- np.mean(a)
```

## One Sample T-Test

Null Hypothesis: mean of a sample is a specific values e.g. Assume for group A it is 50

```python
np.mean(a)
```

```python
t_stat,p_value=stats.ttest_1samp(a,50,axis=0)
```

```python
t_stat,p_value
```

the p-value>0.5 so we accept the null hypotheses

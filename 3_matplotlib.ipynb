```python
brand = ['puma' , 'jockey' , 'nike' , 'levis'  ,'wildcraft' , 'pe'] 
sales = [41, 23,17,15,40,38]
```

```python
plt.pie(sales , labels = brand)
```

```python
plt.pie(sales , labels = brand , startangle = 90)
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1])
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1] , shadow = True)
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1] , colors = ['m' , 'cyan' , 'b' , 'y','r' ,'green'])
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1] , colors = ['m' , 'cyan' , 'b' , 'y','r' ,'green'])
plt.legend()
```

```python
def myf(x) : 
    print(x) 
    return '{:.4f}%\n({:.0f})'.format(x, 174*x/100)
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1] , colors = ['m' , 'cyan' , 'b' , 'y','r' ,'green'],
       autopct=myf)
```

```python
plt.pie(sales , labels = brand ,startangle = 90  , explode = [0.3, 0,0,0,0, 0.1] , colors = ['m' , 'cyan' , 'b' , 'y','r' ,'green'],
       autopct='%.2f')
```
